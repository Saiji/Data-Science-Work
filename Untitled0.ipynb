```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

!pip install streamlit

st.set_page_config(page_title="Demand Forecast Tool", layout="wide")

st.title("📊 Demand Forecast Planning Tool")
st.write("Simple Supply Chain Forecasting App (MA, Exp Smoothing, ML)")

# -----------------------------
# Upload Data
# -----------------------------
uploaded_file = st.file_uploader("Upload Demand Data (CSV)", type=["csv"])

if uploaded_file:
    df = pd.read_csv(uploaded_file)
else:
    st.info("Using sample dataset")
    df = pd.DataFrame({
        "Date": pd.date_range(start="2023-01-01", periods=12, freq='M'),
        "Demand": [120, 135, 150, 160, 170, 180, 175, 190, 200, 210, 220, 230]
    })

# -----------------------------
# Data Prep
# -----------------------------
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")
df.set_index("Date", inplace=True)

st.subheader("📂 Data Preview")
st.dataframe(df)

# -----------------------------
# Parameters
# -----------------------------
st.sidebar.header("⚙️ Settings")

window = st.sidebar.slider("Moving Average Window", 2, 6, 3)
alpha = st.sidebar.slider("Exponential Smoothing Alpha", 0.1, 0.9, 0.3)

# -----------------------------
# Forecast Calculations
# -----------------------------
df["MA_Forecast"] = df["Demand"].rolling(window=window).mean()

df["Exp_Forecast"] = df["Demand"].copy()
for i in range(1, len(df)):
    df.iloc[i, df.columns.get_loc("Exp_Forecast")] = (
        alpha * df["Demand"].iloc[i-1] +
        (1 - alpha) * df["Exp_Forecast"].iloc[i-1]
    )

# -----------------------------
# ML Forecast
# -----------------------------
df["Lag1"] = df["Demand"].shift(1)
df["Lag2"] = df["Demand"].shift(2)

df_ml = df.dropna()

ml_forecast = None
if len(df_ml) > 5:
    X = df_ml[["Lag1", "Lag2"]]
    y = df_ml["Demand"]

    model = RandomForestRegressor()
    model.fit(X, y)

    ml_forecast = model.predict([[df["Demand"].iloc[-1], df["Demand"].iloc[-2]]])[0]

# -----------------------------
# Next Forecast
# -----------------------------
next_ma = df["Demand"].tail(window).mean()
next_exp = alpha * df["Demand"].iloc[-1] + (1 - alpha) * df["Exp_Forecast"].iloc[-1]

# -----------------------------
# Results
# -----------------------------
st.subheader("📈 Forecast Results")

col1, col2, col3 = st.columns(3)

col1.metric("Moving Avg Forecast", f"{next_ma:.2f}")
col2.metric("Exp Smoothing Forecast", f"{next_exp:.2f}")
col3.metric("ML Forecast", f"{ml_forecast:.2f}" if ml_forecast else "Not enough data")

# -----------------------------
# Visualization
# -----------------------------
st.subheader("📊 Forecast vs Actual")

fig, ax = plt.subplots()

ax.plot(df.index, df["Demand"], label="Actual")
ax.plot(df.index, df["MA_Forecast"], label="Moving Avg")
ax.plot(df.index, df["Exp_Forecast"], label="Exp Smoothing")

ax.legend()
plt.xticks(rotation=45)

st.pyplot(fig)

# -----------------------------
# Download Output
# -----------------------------
st.subheader("⬇️ Download Results")

csv = df.to_csv().encode('utf-8')
st.download_button("Download Forecast Data", csv, "forecast_output.csv", "text/csv")
```
